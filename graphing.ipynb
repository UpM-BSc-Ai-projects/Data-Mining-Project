```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('data\preprocessed_hourly_data.csv')

# Ensure the Timestamp column is in datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Set the Timestamp as the index
data.set_index('Timestamp', inplace=True)

# Plot 1: Time Series Line Plot (e.g., Close prices)
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Close', color='blue')
plt.title('Time Series Line Plot (Close Prices)')
plt.xlabel('Timestamp')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()

# Plot 2: Histogram (e.g., Close prices)
plt.figure(figsize=(10, 6))
plt.hist(data['Close'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.grid()
plt.show()
# Plot 3: Scatter Plot (Close Prices vs ATR_168)
plt.figure(figsize=(12, 6))
plt.scatter(data['Close'], data['ATR_168'], alpha=0.5, color='purple')
plt.title('Scatter Plot of Close Prices vs ATR_168')
plt.xlabel('Close Price')
plt.ylabel('ATR_168')
plt.grid()
plt.show()
```
